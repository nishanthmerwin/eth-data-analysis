```python
import json
import seaborn as sns
from tqdm import tqdm
from matplotlib import pyplot as plt
from nmutils import generalutils
import numpy as np
```

```python
with open("../data/bunny_wallets.json") as fp:
    bunny_wallets = json.load(fp)

with open("../data/bat_wallets.json") as fp:
    bat_wallets = json.load(fp)
    
```

```python
print len(bunny_wallets)
print len(bat_wallets)

bat_wallets = [x for x in bat_wallets if x['balance'] > 0]


for wallet in tqdm(bunny_wallets + bat_wallets):    
    txns = wallet['inputs'] + wallet['outputs']
    wallet['num_txns'] = len(txns)
```

```python


fig, ax = plt.subplots(1,1, figsize=(15,8))

sns.kdeplot([x['num_txns'] for x in bunny_wallets if x['num_txns'] < 300],
            color="red", shade=True, label="Bunny", ax=ax,bw=.4, gridsize=2000)
sns.kdeplot([x['num_txns'] for x in bat_wallets if x['num_txns'] < 300], shade=True, 
            color="blue", label="BAT", ax=ax,bw=.4, gridsize=2000)

plt.xlim(0, 20)
#plt.xlim(0, 10)
plt.legend();
plt.xlabel("Number of transactions per wallet")
plt.savefig("../final_vis/transactions.svg")

```
