```python
import json
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
with open("../data/bunny_wallets.json") as fp:
    wallets = json.load(fp)
```

```python
print len(wallets)
wallets = [x for x in wallets if x['balance'] > 0]
print len(wallets)
```

```python
for wallet in wallets:
    wallet['num_transactions'] = len(wallet['inputs']) + len(wallet['outputs'])
```

```python
sns.kdeplot([x['balance'] for x in wallets], shade=True)

#plt.xlim(0, 100000000)
plt.show()
```

```python
sns.kdeplot([x['num_transactions'] for x in wallets], shade=True)

#plt.xlim(0, 10)
plt.show()
```
